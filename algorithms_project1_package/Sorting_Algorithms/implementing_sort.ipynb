```python
'''
Todo:
    quick
    heap
    merge
    mystery
'''

# use as an Enum or Set to check for which id to sort
column_names= ['tconst', 'primaryTitle', 'originalTitle', 'startYear',
               'runtimeMinutes', 'genres', 'averageRating', 'numVotes', 'ordering',
               'category', 'job', 'seasonNumber', 'episodeNumber', 'primaryName', 'birthYear',
               'deathYear', 'primaryProfession']

######################## SORTING-ALGORITHMS ########################

#columns: a list of integers representing the columns to sort the 2D array on

def get_col_indx(columns):
    list = [] 
    for i in columns: list.append(column_names.index(i))
    return list

# used whenever the 
# If first element is less than second, return False, else True.
# If they are equal return Ture
def is_greater(d1, d2, columns):
    cur_idx = 1
    while(True):
        try:
            if d1[columns[cur_idx]] < d2[columns[cur_idx]]:
                return False
            elif d1[columns[cur_idx]] == d2[columns[cur_idx]]:
                cur_idx += 1
            else:
                return True
        except:
            return True

############################# Done ##################################
# Quick Sort
# https://www.geeksforgeeks.org/python-program-for-quicksort/
# based of ^
def pivot_element(arr, low, high, columns):
    pivot = arr[high]
    i = low-1

    for j in range(low, high):
        if arr[j][columns[0]] <= pivot[columns[0]]:
            if arr[j][columns[0]] == pivot[columns[0]]:
                if is_greater(arr[j], pivot, columns): continue
            i = i+1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]

    return i+1

def quicksort(arr, low, high, columns):

    if low < high:
        # Find pivot element such that
        # element smaller than pivot are on the left
        # element greater than pivot are on the right
        pi = pivot_element(arr, low, high, columns)
        # Recursive call on the left of pivot
        quicksort(arr, low, pi - 1, columns)
        # Recursive call on the right of pivot
        quicksort(arr, pi + 1, high, columns)
############################# Done ##################################

############################# Done ##################################
# Selection Sort
def selection_sort(arr, columns):
    n = len(arr)
    
    # 2d - arr, always increase by one 
    for i in range(0, n):
        min_idx = i
        for j in range(i+1, n):

            if arr[j][columns[0]] <= arr[min_idx][columns[0]]:

                if arr[j][columns[0]] == arr[min_idx][columns[0]]:
                    if is_greater(arr[j], arr[min_idx], columns): continue
                min_idx = j
            
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return arr
############################# Done ##################################


#Heap Sort
def max_heapify(arr, n, i, columns):
    pass

def build_max_heap(arr, n, i, columns):
    pass

def heap_sort(arr, columns):
    return arr

############################# Done ##################################
# https://www.geeksforgeeks.org/python-program-for-shellsort/
# based from ^
# Shell Sort
def shell_sort(arr, columns):
    n = len(arr)
    gap = int(n/2)
 
    while gap > 0:
 
        for i in range(gap, n):
 
            key = arr[i]
            j = i
            while j >= gap and key[columns[0]] <= arr[j-gap][columns[0]]:
                if key[columns[0]] == arr[j-gap][columns[0]]:
                    if is_greater(key, arr[j-gap], columns): break

                arr[j] = arr[j-gap]
                j -= gap
 
            arr[j] = key
        gap /= 2
        gap = int(gap)
    
    return arr
############################# Done ##################################

#Merge Sort
def merge(left, right, columns):
    pass

def merge_sort(data, columns):
    # if len(data) <= 1:
    #     return data
    # return merge(left, right, columns)
    pass



############################# Done ##################################
# https://www.geeksforgeeks.org/python-program-for-insertion-sort/
# based of ^
# insertion_sort
def insertion_sort(arr, columns):
    n = len(arr)
    for i in range(1, n):
        
        key = arr[i]
        j = i-1
        while j >= 0 and key[columns[0]] <= arr[j][columns[0]]:
                if key[columns[0]] == arr[j][columns[0]]:
                    if is_greater(key, arr[j], columns): break
                arr[j+1] = arr[j]
                j -= 1
        arr[j+1] = key
    
    return arr
############################# Done ##################################

```
